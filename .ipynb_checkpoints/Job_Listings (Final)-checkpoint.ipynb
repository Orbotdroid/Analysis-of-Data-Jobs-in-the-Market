```python
# Set Executable Path & Initialize Chrome Browser
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Parse Results HTML with BeautifulSoup
html = browser.html
soup_splinter = bs(html, 'html.parser')
```

```python
li_job_pages = []

def li_scraper(keyword, location, pages):
# Convert All Keyword Strings with Spaces Into URL Format with "%20"
    url_keyword = []
    for i in range(0, len(keyword)):
        if keyword[i] == " ":
            url_keyword.append("%20")
        else: url_keyword.append(keyword[i])
        key_string = ''.join(url_keyword)
    
# Convert All Location Strings with Spaces Into URL Format with "%20"
    url_location = []
    for i in range(0, len(location)):
        if location[i] == " ":
            url_location.append("%20")
        else: url_location.append(location[i])
        loc_string = ''.join(url_location)
        
# Generate URL
    for i in range(0,int(pages)):
        url = f"https://www.linkedin.com/jobs/search/?keywords={key_string}&location={loc_string}%2C%20California&locationId=PLACES.us.7-1-0-38-1&start=" + str(i*25)

        # Scraping URL
        browser.visit(url)
        html = browser.html
        soup_splinter = bs(html, 'html.parser')
        li_job_pages.append(soup_splinter)

    # Variables that Store Their Associated Category of Data (Tags Included)
    job_titles = []
    for i in range(0, len(li_job_pages)):
        job_titles.append(li_job_pages[i].find_all("h3", attrs={"class": "listed-job-posting__title",
                                   }))
    job_locations = []
    for i in range(0, len(li_job_pages)):
        job_locations.append(li_job_pages[i].find_all("p", attrs={"class":"listed-job-posting__location"}))
    
    job_companies = []
    for i in range(0, len(li_job_pages)):
        job_companies.append(li_job_pages[i].find_all("h4", attrs={"class": "listed-job-posting__company"}))
    
    job_time_posted_ago = []
    for i in range(0, len(li_job_pages)):
        job_time_posted_ago.append(li_job_pages[i].find_all("span", attrs={"class":"listed-job-posting__flavor posted-time-ago__text"}))

    # Variables that Contain the Same Data (Tags Removed)
    locations = []
    for i in range(0, len(job_locations)):
        for j in range(0, len(job_locations[i])):
            locations.append(job_locations[i][j].get_text())
    
    time_posted = []
    for i in range(0, len(job_time_posted_ago)):
        for j in range(0, len(job_time_posted_ago[i])):
            time_posted.append(job_time_posted_ago[i][j].get_text())
    
    companies = []
    for i in range(0, len(job_companies)):
        for j in range(0, len(job_time_posted_ago[i])):
            companies.append(job_companies[i][j].get_text())
    
    titles = []
    for i in range(0, len(job_titles)):
        for j in range(0, len(job_titles[i])):
            titles.append(job_titles[i][j].get_text())

    # Return Time Posted, Companies, Job Titles & Locations
    return time_posted, companies, titles, locations
```

```python
# Web Scrape 20 Pages of Job Listings on LinkedIn
job_listings = li_scraper("data analyst", "san francisco", 20)
```

```python
# Extract Only Company Name to Search Google Maps API for GeoCoordinates
company = job_listings[["Company Name"]]
company.head()
```

```python
# Find the GeoCoordinates (Latitude/Longitude) of the Companies Scraped 
company_list = list(company["Company Name"])

# Build URL Using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"
new_json = []

for target_company in company_list:

    params = {"address": target_company + ", San Francisco CA", "key": gkey}

    # Run Request
    response = requests.get(base_url, params=params)
    
    # Error Handling to Bypass IndexError's
    try:
        # Extract lat/lng
        companies_geo = response.json()
        lat = companies_geo["results"][0]["geometry"]["location"]["lat"]
        lng = companies_geo["results"][0]["geometry"]["location"]["lng"]
        new_json.append({"company": target_company, "lat": lat, "lng": lng})
        places = f"{target_company}, {lat}, {lng}"
        print(places)
    except IndexError:
        print("Error at ", places)
```

```python
print(new_json)
```

```python
# Convert JSON into GeoJSON
geojson = {
    "type": "FeatureCollection",
    "features": [
    {
        "type": "Feature",
        "company": d["company"],
        "geometry" : {
            "type": "Point",
            "coordinates": [d["lat"], d["lng"]],
            },
     } for d in new_json]
}
print(geojson)
```

```python
# Convert to JSON & Save JSON File
with open('data.json', 'w') as outfile:
    outfile.write(str(geojson))
```
