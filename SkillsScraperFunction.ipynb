# Data Mungling

```python
# Function that Arrange > Cleans > Formating All the Data

def dataMungling (list1, list2):

    # parsing profile column into a Data frame
    data_position = pd.DataFrame(data=[profile for profile in list1], columns=["profile"])

    # parsing skill columns
    data_position["skill"] = pd.DataFrame(data=[profile for profile in list2], columns=['skills'])


    # Checking the syntax to execute the mungling
    #data_position.iloc[1,1]
    #data_position.iloc[1,0]
    #data_position.head()


    data_position['len'] = np.array([len(profile) for profile in list1])
    data_position["name"] = ""
    data_position["position"] = ""
    data_position["company"] = ""
    data_position["trash"] = ""


    print(data_position["profile"][1].split(" - "))

    for x in range(len(data_position["profile"])):
            
        d = data_position["profile"][x].split(" - ")
            #print(d)
            
        try:
            data_position["name"][x] = d[0]
        except:
             print("error")

        try:
            data_position["position"][x] = d[1]
        except: 
            print("error")

        try:
            data_position["company"][x] = d[2]
        except:
            print("error")

    #data_position.head()


    # For the Articles

    data_position["trash2"] = ""
    data_position["city"] = ""


    for x in range(len(data_position["skill"])):
        try:
            d = data_position["skill"][x].split(" ... ")
        except:
            print("error")
            #print(d)
            
            
        try:
            data_position["trash2"][x] = d[0]
        except:
             print("error")

        try:
            data_position["city"][x] = d[1]
        except: 
            print("error")

        try:
            data_position["skill"][x] = d[2]
        except:
            print("error")


    del data_position["trash"]
    del data_position["trash2"]
    del data_position["profile"]


    
    return data_position.head(100)

    # Transform into csv

    # engine = create_engine('sqlite://',echo=False)
    # data_position.to_csv(r'C:\Users\Piter\Desktop\Berkeley BC\Week-12-Web-Scraping\DataPosition.csv')

dataMungling(analyst_title, analyst_article)    
```
